### Extracting Training Data

```python
X=car[['name','company','year','kms_driven','fuel_type']]
y=car['Price']
```

### Applying Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

#### Creating an OneHotEncoder object to contain all the possible categories

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

#### Creating a column transformer to transform categorical columns

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

#### Linear Regression Model

```python
lr=LinearRegression()
```

#### Making a pipeline

```python
pipe=make_pipeline(column_trans,lr)
```

#### Fitting the  model

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

#### Checking R2 Score

```python
r2_score(y_test,y_pred)
```

#### Finding the model with a random state of TrainTestSplit where the model was found to give almost 0.92 as r2_score

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
pipe.predict(pd.DataFrame(columns=X_test.columns,data=np.array(['Maruti Suzuki Swift','Maruti',2019,100,'Petrol']).reshape(1,5)))
```

#### The best model is found at a certain random state

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```

Basic interactive prediction


```python
from ipywidgets import interactive, Text, IntSlider, Dropdown, Button, VBox, Output
from IPython.display import display
import pandas as pd
import numpy as np

# Assuming your 'pipe' and 'X_test.columns' are available from your training code
# and that your 'pipe' has already been fitted.

# Access the fitted OneHotEncoder from the fitted pipeline
fitted_ohe = pipe.named_steps['columntransformer'].transformers_[0][1]

# Get the categories after fitting
name_categories = fitted_ohe.categories_[0]
company_categories = fitted_ohe.categories_[1]
fuel_type_categories = fitted_ohe.categories_[2]


# Create input widgets
name_widget = Dropdown(options=name_categories, description='Name:')
company_widget = Dropdown(options=company_categories, description='Company:')
year_widget = Text(description='Year:')  # Changed to Text widget
kms_driven_widget = Text(description='Kms Driven:')  # Changed to Text widget
fuel_type_widget = Dropdown(options=fuel_type_categories, description='Fuel Type:')


predict_button = Button(description="Predict Price")
output = Output()

def predict_price(name, company, year, kms_driven, fuel_type):
    # Convert year and kms_driven to appropriate types (int)
    try:
        year = int(year)
        kms_driven = int(kms_driven)
    except ValueError:
        with output:
            output.clear_output()
            print("Please enter valid numerical values for Year and Kms Driven.")
        return

    # Create a DataFrame in the expected format
    input_data = pd.DataFrame([[name, company, year, kms_driven, fuel_type]],
                              columns=X_test.columns)

    # Make the prediction
    predicted_price = pipe.predict(input_data)[0]

    with output:
        output.clear_output()
        print(f"Predicted Price: {predicted_price:.2f}")

# Link the button click to the prediction function
predict_button.on_click(lambda b: predict_price(name_widget.value, company_widget.value,
                                                year_widget.value, kms_driven_widget.value,
                                                fuel_type_widget.value))

# Arrange widgets in a vertical box
ui = VBox([name_widget, company_widget, year_widget, kms_driven_widget, fuel_type_widget, predict_button, output])

# Display the UI
display(ui)
```
